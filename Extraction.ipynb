```python
import geopandas as gpd
def extract(input_file, output_name):
    """
    Extracts attribute data from a shapefile and saves it as an Excel file.
    
    Parameters:
        input_file (str): Path to the input .shp file
        output_name (str): Desired output Excel file name (without extension)
    """
    # Load shapefile
    gdf = gpd.read_file(input_file)
    
    # Drop geometry (keep only tabular data)
    table = gdf.drop(columns='geometry')
    
    # Save to Excel using the output name
    table.to_excel(f"{output_name}.xlsx", index=False)
```

```python
downloaded_files = [(r"autoctonia_2022\PctAutoc_2022.shp","auctoctonia_2022"),
                   (r"densidad_poblacion_2023\DensPob2023.shp","densidad_poblacion_2023"),
                   (r"edad_media_2023\EdadMedia2023.shp", "edad_media_2023"),
                   (r"num_afiliados-1000_2022\NumAfi_2022.shp", "num_afiliados-1000_2022"),
                   (r"num_parados-1000_2022\NumParados_2022.shp", "num_parados-1000_2022"),
                   (r"poblacion-estudiossup16_2021\PctPob16_2021.shp", "poblacion-estudiossup16_2021"),
                   (r"poblacion-extranjera_2022\PctPobExt_2022.shp","poblacion-extranjera_2022"),
                   (r"renta_media_neta_hogar_2020\RentaMedia_2020.shp", "renta_media_neta_hogar_2020")]
```

```python
for input_file, output_name in downloaded_files:
    extract(input_file,output_name)
```
